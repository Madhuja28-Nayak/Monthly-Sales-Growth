```python
import pandas as pd
```

```python
df=pd.read_excel("sales_raw_500.xlsx")
```

```python
print(df)
```

```python
df=df.drop_duplicates()
print(df)
```

```python
#fix date column(to remove cells in the columns with null value)
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'],errors='coerce')
```

```python
df=df.dropna(subset=['Order_Date'])
```

```python
print(df)
```

```python
#to extract month and year along with adding profit column
```

```python
df['Profit']=df['Sales']-df['Cost']
df['Year']=df['Order_Date'].dt.year
df['Month']=df['Order_Date'].dt.month
```

```python
print(df)
```

```python
#to remove the index column and convert data into csv format 
```

```python
df.to_csv('sales_cleaned.csv', index=False)
```
